# Combine Filters

```python
# Ideas
# First, convert the image into gray scale image
# Second, find edge of the image
# Last, sharpen the image

from PIL import Image, ImageFilter
import numpy as np
from matplotlib import pyplot as plt

img = Image.open("images/eye_1.jpg")

img_L = img.convert("L") # 
img_COUNTER = img_L.filter(ImageFilter.CONTOUR) # 
img_SHARPEN = img_COUNTER.filter(ImageFilter.SHARPEN) # 

plt.imshow(img_SHARPEN, plt.cm.gray)
plt.show()
```

# Use OpenCV

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('images/eye_1.jpg',0)
edges = cv2.Canny(img, 100, 200)

plt.subplot(121), plt.imshow(img, cmap='gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges, cmap='gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
plt.imshow(edges, cmap='gray')
plt.show()
```

```python
img = cv2.imread('images/eye_1.jpg',0)

# sharpen the image
kernel = np.array([[-1,-1,-1], [-1,9,-1], [-1,-1,-1]]) # sharpen kernel
img_sharpen = cv2.filter2D(img, -1, kernel)

# find edge
edges = cv2.Canny(img_sharpen, 100, 200)

plt.imshow(edges, cmap='gray')
plt.show()
```
